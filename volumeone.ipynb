<font size="3"><b> Type of Degrees Based on Researchers<b></font>
    
All of the researchers from volume one have earned mulitple degrees based on their respected discplines. The chart below shows three different type of degrees: bachelors, masters, and ph.D. 
    
The bar chart called Researchers Level of Education shows the value for each of the degree:
<p>·ph.D = 11 <br>
·Masters = 8 <br>
·Bachelors = 11 <p>


```python
import pandas as pd
from matplotlib import pyplot as plt
 
# Read CSV into pandas
data = pd.read_csv(r"vone.csv")
data.head()
df = pd.DataFrame(data)

# Figure Size for plot
fig = plt.figure(figsize = (10, 5))
 
# Horizontal Bar Plot for Education
df['Education'].value_counts().plot(kind='bar', color = '#104E8B', rot=0, width = 0.4) 

plt.xlabel("\n Type of Degrees")
plt.ylabel("No. of Degrees")
plt.title("Researchers Level of Education")

# Displays Plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vone.csv')
colors = ["#1f77b4", "#ff7f0e"]

# edit the CSV for this chart
group_by_diag = df.groupby("Research Method").count().reset_index()
sizes = group_by_diag['Research Volume'] # insert numbers 
labels = group_by_diag['Research Method']

fig = plt.figure(figsize = (8, 8))

plt.pie(sizes, labels = labels, colors = colors)
plt.title('Research Method Type')
plt.show()
```
